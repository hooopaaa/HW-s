# Задание 1. Часть 1

```python
class AnimalNotFoundError(Exception):
    pass

class Animal:
    _all_animals = []

    def __init__(self, name, height):
        self.name = name
        self.height = height
        Animal._all_animals.append(self)

    def speak(self):
        print("...")

    @classmethod
    def get_highest(cls):
        if not cls._all_animals:
            raise AnimalNotFoundError("Животные не найдены")
        return max(animal.height for animal in cls._all_animals)

    @classmethod
    def get_highest_animal(cls):
        if cls._all_animals:
            return max(cls._all_animals, key=lambda animal: animal.height)
        return None


a1 = Animal("Alex", 200)
a2 = Animal("Melman", 400)
a3 = Animal("King Julien XIII", 40)
a4 = Animal("Maurice", 40)

try:
    highest = Animal.get_highest()
except AnimalNotFoundError as e:
    print(e)
else:
    print(f"Самый высокий рост среди животных: {highest}")

    assert Animal.get_highest() == 400

    highest_animal = Animal.get_highest_animal()
    if highest_animal:
        print(f"Самое высокое животное это {highest_animal.name}, его рост целых {highest_animal.height}")
```

```python
class Animal:
    _all_animals = []

    def __init__(self, name, height):
        self.name = name
        self.height = height
        Animal._all_animals.append(self)

    def speak(self):
        print("...")

    @classmethod
    def get_highest(cls):
        if not cls._all_animals:
            raise ValueError("Животные не найдены")
        return max(animal.height for animal in cls._all_animals)

    @classmethod
    def get_highest_animal(cls):
        if cls._all_animals:
            return max(cls._all_animals, key=lambda animal: animal.height)
        return None

    def __eq__(self, other):
        # Возвращаем True только если объекты одного и того же класса
        return isinstance(other, self.__class__)


class Lion(Animal):
    pass

class Lemur(Animal):
    pass

class Giraffe(Animal):
    pass


# Создание экземпляров животных разных видов
a1 = Lion("Alex", 200)
a2 = Giraffe("Melman", 400)
a3 = Lemur("King Julien XIII", 40)
a4 = Lemur("Maurice", 40)

# Проверка сравнений и вывод результатов
print(f"Сравнение a4 и a3 (оба Lemur): {a4 == a3}")
assert a4 == a3

print(f"Сравнение a1 и a2 (Lion и Giraffe): {a1 != a2}")  # Ожидается True
assert a1 != a2

print(f"Сравнение a2 и a4 (Giraffe и Lemur): {a2 != a4}")  # Ожидается True
assert a2 != a4

print("Проверки прошли успешно.")
```

```python
import unittest
from contextlib import contextmanager

class Animal:
    _all_animals = []

    def __init__(self, name, height):
        self.name = name
        self.height = height
        Animal._all_animals.append(self)

    def speak(self):
        raise NotImplementedError("Каждый класс должен реализовать метод speak.")

    @classmethod
    def get_highest(cls):
        if not cls._all_animals:
            raise ValueError("Животные не найдены")
        return max(animal.height for animal in cls._all_animals)

    @classmethod
    def get_highest_animal(cls):
        if cls._all_animals:
            return max(cls._all_animals, key=lambda animal: animal.height)
        return None


class Lion(Animal):
    def speak(self):
        print("Roar")

class Lemur(Animal):
    def speak(self):
        print("I'm a Lemuuuur")

class Giraffe(Animal):
    def speak(self):
        print("oh hi")


class AnimalTestCase(unittest.TestCase):
    @contextmanager
    def assertNotRaises(self, exc_type):
        try:
            yield None
        except exc_type:
            raise self.failureException(f'{exc_type.__name__} raised')

    def test_create_animal_instance(self):
        a = Animal("Generic Animal", 100)
        with self.assertRaises(NotImplementedError):
            a.speak()

    def test_create_zebra(self):
        with self.assertRaises(NotImplementedError):
            class Zebra(Animal):
                pass
            a = Zebra("Marty", 200)
            a.speak()

    def test_create_animals(self):
        with self.assertNotRaises(TypeError):
            a1 = Lion("Alex", 200)
            a2 = Giraffe("Melman", 400)
            a3 = Lemur("King Julien XIII", 40)
            a4 = Lemur("Maurice", 40)
            a1.speak()
            a2.speak()
            a3.speak()
            a4.speak()

unittest.main(argv=[''], verbosity=0, exit=False)
```
